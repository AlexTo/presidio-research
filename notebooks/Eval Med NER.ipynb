```python
import spacy
from presidio_evaluator.data_generator import read_synth_dataset
from presidio_evaluator import SpacyEvaluator
```

```python
# nlp = spacy.load("../models/output/model-best/")
nlp = spacy.load("../../nlp-assets/modules/tag_entities/models/en/nlp")
```

```python
val_data = read_synth_dataset('/data/datasets/n2c2/2012/processed/test.json')
```

```python
evaluator = SpacyEvaluator(model=nlp, labeling_scheme="BILOU", translate_to_spacy_ents=False)
```

```python
results = evaluator.evaluate_all(val_data)
```

```python
results = evaluator.calculate_score(results)
```
